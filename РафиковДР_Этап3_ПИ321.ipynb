**Генерируем данные**

```python
# Загружаем библиотеку numpy
import numpy as np
```

```python
# Номер зачётки - начальное состояние генератора
np.random.seed(19130156)

# Среднее значение (математическое ожидание)
mu = 100

# Сигма (стандартное отклонение)
sigma = 10

# Размер (объём) выборки
N = 10000

# Генератор случайных чисел с нормальным распределением
x = np.random.normal(mu, sigma, N)
```

```python
# Загружаем графическую библиотеку
import matplotlib.pyplot as plt
```

***Гистограмма (столбиковая диаграмма частот)***

```python
# Подбираем число интервалов группировки
n = 10

# Гистограмма распределения
plt.hist(x, n)
```

```python
# Размеры графика в дюймах
plt.figure(figsize=(13, 5))
# Цвет и толщина линий, цвет заливки столбиков
plt.hist(x, bins=n, edgecolor='black', linewidth=3, facecolor='white', density=True)
# Заголовок графика
plt.title("Гистограмма нормального распределения")
# Не выводить результаты группировки данных на экран
plt.show()
```

```python
# Загружаем нормальное распределение
from scipy.stats import norm

# Массив заданного количества "иксов" для графика
xx = np.linspace(50, 150, 100)

# График плотности вероятности нормального распределения
plt.plot(xx, norm.pdf(xx, loc=mu, scale=sigma))
plt.show()
```

```python
plt.figure(figsize=(13, 5))
plt.hist(x, bins=n, edgecolor='black', linewidth=3, facecolor='white', density=True, label='Гистограмма')
plt.plot(xx, norm.pdf(xx, loc=mu, scale=sigma), color="blue", linewidth=3, label="Нормальное распределение")
plt.title("Нормальное распределение")
plt.xlabel("Значение случайной величины")
plt.ylabel("Плотность вероятности")
plt.legend()
plt.text(x=50, y=0.03, s=f"$\mu$ = {mu}\n$\sigma$ = {sigma}")
plt.show()
```

***Кумулята (график накопленных частот)***

```python
plt.hist(x, bins=n, density=True, cumulative=True)
```

```python
(C_cum, x_cum, p_cum) = plt.hist(x, bins=n, density=True, cumulative=True)
```

```python
x_cum.shape
```

```python
C_cum.shape
```

```python
plt.plot(x_cum[1:], C_cum)
```

```python
# Кумулята и функция нормального распределения
plt.figure(figsize=(13, 5))
plt.plot(x_cum[1:], C_cum, color='black', linewidth=3, label='Кумулята', marker='o')
С = norm.cdf(xx, loc=mu, scale=sigma)
plt.plot(xx, С, color="blue", linewidth=3, label="Нормальное распределение")
plt.title("Нормальное распределение")
plt.xlabel("Значение случайной величины")
plt.ylabel("Вероятность (накопленные частоты)")
plt.legend()
plt.text(x=55, y=0.8, s=f"$\mu$ = {mu}\n$\sigma$ = {sigma}")
plt.show()
```
