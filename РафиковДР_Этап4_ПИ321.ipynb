> Выполнил Рафиков Д.Р. ПИ321 Этап 4



````python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

#номер зачетки
np.random.seed(19130156) 

#Объем выборки
N=100

#Генерация
x=np.random.uniform(low=150,high=200,size=N)
np.random.uniform(low=0.0,high=1.0,size=None)

#Обратный номер зачетки
np.random.seed(65103191)

e=np.random.normal(loc=0,scale=13,size=N)
np.random.uniform(low=0.0,high=1.0,size=None)
y=-100+x+e

#Рисование
plt.figure(figsize=(13,13))
plt.scatter(x,y)
plt.title("Диаграмма разброса")
plt.xlabel("x")
plt.ylabel("y")

r=st.pearsonr(x,y)[0]
rho=st.spearmanr(x,y)[0]
tau=st.kendalltau(x,y)[0]

#Отладка
print(f"r Пирсона: \t{r:.3f}")
print(f"rho Спирмана: \t{rho:.3f}")
print(f"tau Кендалла: \t{tau:.3f}")

results_yx=st.linregress(x,y)
a=results_yx.intercept
b=results_yx.slope
a,b

print(f"a: \t{a:.3f}")
print(f"b: \t{b:.3f}")

plt.scatter(x,y, label="Исходные данные",color = '#667777')
x_yx=np.array([150,200])
y_yx=a+b*x_yx

plt.plot(x_yx,y_yx, color="brown", label='Линия регрессии y(x)')
results_xy=st.linregress(y,x)
c=results_xy.intercept
d=results_xy.slope
c,d
print(f"c: \t{c:.3f}")
print(f"d: \t{d:.3f}")

x_mean=np.mean(x)
y_mean=np.mean(y)
x_mean,y_mean
print(f"x_mean: \t{x_mean:.3f}")
print(f"y_mean: \t{y_mean:.3f}")

y_yx=np.array([35,120])
x_yx=c+d*y_yx

plt.plot(x_yx,y_yx,color="blue",label="Линия регрессии x(y)")
plt.scatter(x_mean,y_mean, color="red",label="Средние значения")
plt.legend()
plt.show()
````
